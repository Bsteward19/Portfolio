```python
import pandas as pd
import numpy as np
Test_data_Target = pd.read_csv('C:\\Users\\brend\\Documents\\Year 3\\Dissertation\\Data\\Test_Data.csv', encoding='utf8')
df = pd.read_csv('C:\\Users\\brend\\Documents\\Year 3\\Dissertation\\Data\\swift_dataset (1).csv', encoding='windows-1254')
df['Abstract'].fillna(0, inplace=True)
abstracts = list(df['Abstract'])
labels = list(df['Included (Full Text)']) 
Test_labels = list(Test_data_Target['Included (Full Text)'])
abstracts = [str (item) for item in abstracts]
Test_abstract = list(Test_data_Target['Abstract'])
Test_abstract = [str (item) for item in Test_abstract]
print("Inputted Datasets")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(abstracts)
Y = vectorizer.transform(Test_abstract)
print(vectorizer.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X)
X_train_tfidf.shape
print(X_train_tfidf)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import numpy as np
from sklearn.metrics import confusion_matrix
clf = MultinomialNB().fit(X_train_tfidf, labels)
print("Done Training")
predicted = clf.predict(Y)
print('done predicting')
print(np.mean(predicted == Test_labels))
print(classification_report(Test_labels, predicted, digits = 3))
```
