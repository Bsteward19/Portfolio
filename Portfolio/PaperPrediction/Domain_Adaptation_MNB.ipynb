```python
import pandas as pd
import numpy as np
Test_data_Target = pd.read_csv('C:\\Users\\brend\\Documents\\Year 3\\Dissertation\\Data\\Test_Data.csv', encoding='utf8')
df = pd.read_csv('C:\\Users\\brend\\Documents\\Year 3\\Dissertation\\Data\\Target_Data.csv', encoding='windows-1254')
df['Abstract'].fillna(0, inplace=True) #replace blanks with 0
abstracts = list(df['Abstract']) #put "Abstract" column in list
labels = list(df['Included (Full Text)']) #put "Included (Full Text) column in list
Test_labels = list(Test_data_Target['Included (Full Text)'])
abstracts = [str (item) for item in abstracts] #convert objects in abstracts array to string
Test_abstract = list(Test_data_Target['Abstract'])
Test_abstract = [str (item) for item in Test_abstract]
print("Inputted Datasets")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(abstracts) #tokenise words in CountVectorizer list and transforms into feature vectors
Y = vectorizer.transform(Test_abstract)
print(vectorizer.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X) #transform into tfidf so more common words have similar weights to less common
X_train_tfidf.shape
print(X_train_tfidf)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
import numpy as np
from sklearn.metrics import confusion_matrix
clf = MultinomialNB().fit(X_train_tfidf, labels) #fit onto algorithm
print("Done Training")
predicted = clf.predict(Y) #predict Test Data
print('done predicting')
print(np.mean(predicted == Test_labels)) #Compare Test Data results with actual labels
print(classification_report(Test_labels, predicted, digits = 3)) #print report detailing accuracy of algorithm
```
